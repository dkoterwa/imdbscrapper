```python
import pandas as pd
import numpy as np

import requests
from requests import get
from bs4 import BeautifulSoup

from time import sleep
from random import randint
```

```python
# Getting English translated titles from the movies
headers = {'Accept-Language': 'en-US, en;q=0.5'}
```

```python
pages = np.arange(1, 12051, 50)
pages
```

```python
titles = []
titles_keys = []
years = []
time = []
```

```python
# Storing each of the urls of 50 movies 
from tqdm import tqdm
def base_info(url):
    for page in tqdm(pages):
        # Getting the contents from the each url
        page = requests.get(url + str(page) + '&ref_=adv_nxt', headers=headers)
        soup = BeautifulSoup(page.text, 'html.parser')
        
        # Aiming the part of the html we want to get the information from
        movie_div = soup.find_all('div', class_='lister-item mode-advanced')
        
        # Controling the loopâ€™s rate by pausing the execution of the loop for a specified amount of time
        # Waiting time between requests for a number between 2-10 seconds
        sleep(randint(2,10))
        
        for container in movie_div:

            #Scraping the movies individual key
            key = container.find("div", {"class": "ribbonize", "data-tconst": True})['data-tconst']
            titles_keys.append(key)
            
            # Scraping the movie's name
            name = container.h3.a.text
            titles.append(name)
            
            # Scraping the movie's year
            year = container.h3.find('span', class_='lister-item-year').text
            years.append(year)
            
            # Scraping the movie's length
            runtime = container.find('span', class_='runtime').text if container.p.find('span', class_='runtime') else '-'
            time.append(runtime)
       
```

```python
metascores = []
genres = []
certificates = []

budgets = []
gross_domestic_box_office = []
opening_weekend_box_office = []
cumulative_worldwide_box_office = []
votes = []
ratings = []
popularity_scores = []
oscars_info = []
awards_info = []
release_dates = []
users_reviews_count = []
critics_reviews_count = []
```

```python
from tqdm import tqdm

def main_info(url):
    for title in tqdm(titles_keys):
        page = requests.get(url + title, headers = headers)
        soup = BeautifulSoup(page.text, 'html.parser')
        
        
        box_office_section = soup.find_all('div',attrs={"data-testid": "title-boxoffice-section"})
        
        if len(box_office_section) > 0:
            budget = box_office_section[0].find('li', attrs={"data-testid": "title-boxoffice-budget"}) if box_office_section[0].find('li', attrs={"data-testid": "title-boxoffice-budget"}) else None
            gross_domestic = box_office_section[0].find('li', attrs={"data-testid": "title-boxoffice-grossdomestic"}) if box_office_section[0].find('li', attrs={"data-testid": "title-boxoffice-grossdomestic"}) else None
            opening_weekend_domestic = box_office_section[0].find('li', attrs={"data-testid": "title-boxoffice-openingweekenddomestic"}) if box_office_section[0].find('li', attrs={"data-testid": "title-boxoffice-openingweekenddomestic"}) else None
            cumulative_worldwide_gross = box_office_section[0].find('li', attrs = {"data-testid": "title-boxoffice-cumulativeworldwidegross"}) if box_office_section[0].find('li', attrs = {"data-testid": "title-boxoffice-cumulativeworldwidegross"}) else None


        if budget is not None:
                budgets.append (budget.text)
        else: 
                budgets.append(None)

        if gross_domestic is not None:
                gross_domestic_box_office.append (gross_domestic.text)
        else:
                gross_domestic_box_office.append(None)
            
        if opening_weekend_domestic is not None:
                opening_weekend_box_office.append (opening_weekend_domestic.text)
        else:
                opening_weekend_box_office.append (None)
                
        if cumulative_worldwide_gross is not None:
                cumulative_worldwide_box_office.append (cumulative_worldwide_gross.text)
        else:
            cumulative_worldwide_box_office.append (None)



        votes_number = soup.find('div', attrs = {"class": "sc-7ab21ed2-3 dPVcnq"})
        rating = soup.find('span', attrs = {"class": "sc-7ab21ed2-1 jGRxWM"})
        popularity_score = soup.find('div', attrs = {"class": "sc-edc76a2-1 gopMqI"})

        if votes_number is not None:
            votes.append (votes_number.text)
        else:
            votes.append (None)
        
        if rating is not None:
            ratings.append (rating.text)
        else:
            ratings.append (None)
        
        if popularity_score is not None:
            popularity_scores.append (popularity_score.text)
        else:
            popularity_scores.append (None)

        award_information = soup.find_all('li', attrs = {"data-testid": "award_information"})

        if len(award_information) > 0:
            award_information = soup.find_all('li', attrs = {"data-testid": "award_information"})
            oscars = award_information[0].find(attrs = {"class": "ipc-metadata-list-item__label ipc-metadata-list-item__label--link"}) if award_information[0].find(attrs = {"class": "ipc-metadata-list-item__label ipc-metadata-list-item__label--link"}) else None
            awards = award_information[0].find('span', attrs = {"class": "ipc-metadata-list-item__list-content-item"}) if award_information[0].find('span', attrs = {"class": "ipc-metadata-list-item__list-content-item"}) else None
            oscars_info.append (oscars.text)
            awards_info.append (awards.text)
        else:
            oscars_info.append (None)
            awards_info.append (None)
        

        


        release = soup.find('li', attrs = {"data-testid": "title-details-releasedate"})

        if release is not None:
            release_date = release.find(attrs = {"class": "ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"})
            release_dates.append(release_date.text)
        else:
            release_dates.append(None)



        reviews = soup.find_all('span', attrs = {"class": "score"})

        if len(reviews) == 3:
            users_review = reviews [0].text
            critics_review = reviews [1].text
            metascore = reviews [2].text
        
        elif len(reviews) == 2:
            users_reviews = reviews [0].text
            critics_reviews = reviews [1].text
            metascore = None
        
        elif len(reviews) == 1:
            users_reviews = reviews [0].text
            critics_reviews = None
            metascore = None
        
        else:
            users_reviews = None
            critics_reviews = None
            metascore = None



        users_reviews_count.append (users_review)
        critics_reviews_count.append (critics_review)
        metascores.append (metascore)

        certificate_info = soup.find_all('span', attrs = {'class' : 'sc-8c396aa2-2 itZqyK'})

        if len(certificate_info) == 2:
            certificates.append(certificate_info[1].text)
        else:
            certificates.append(None)

        genres_info = soup.find_all ('li', attrs = {'data-testid' : 'storyline-genres'})

        if len(genres_info) != 0:
            genre = genres_info [0].find(attrs ={'class': 'ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link'})
            genres.append (genre.text)
        else:
            genres.append(None)





```

```python
record = {
    'title_key' : titles_keys,
    'film_title' : titles,
    'year' : years,
    'duration' : time,  
    'budget' : budgets,
    'gross_domestic' : gross_domestic_box_office,
    'opening_weekend' : opening_weekend_box_office,
    'cumulative_box_office' : cumulative_worldwide_box_office,
    'votes' : votes,
    'rating' : ratings,
    'popularity' : popularity_scores,
    'oscars_info' : oscars_info,
    'awards_info' : awards_info,
    'release_date' : release_dates,
    'users_review_count' : users_reviews_count,
    'critics_review_count' : critics_reviews_count,
    'metascore' : metascores,
    'certificate' : certificates,
    'genre' : genres

}
```
